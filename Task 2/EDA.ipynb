# Data Description

### client_data.csv

1. **id** = client company identifier
3. **channel_sales** = code of the sales channel
4. **cons_12m** = electricity consumption of the past 12 months
5. **cons_gas_12m** = gas consumption of the past 12 months
6. **cons_last_month** = electricity consumption of the last month
7. **date_activ** = date of activation of the contract
8. **date_end** = registered date of the end of the contract
9. **date_modif_prod** = date of the last modification of the product
10. **date_renewal** = date of the next contract renewal
11. **forecast_cons_12m** = forecasted electricity consumption for next 12 months
12. **forecast_cons_year** = forecasted electricity consumption for the next calendar year
13. **forecast_discount_energy** = forecasted value of current discount
14. **forecast_meter_rent_12m** = forecasted bill of meter rental for the next 2 months
15. **forecast_price_energy_off_peak** = forecasted energy price for 1st period (off peak)
16. **forecast_price_energy_peak**= forecasted energy price for 2nd period (peak)
17. **forecast_price_pow_off_peak** = forecasted power price for 1st period (off peak)
18. **has_gas** = indicated if client is also a gas client
19. **imp_cons** = current paid consumption
20. **margin_gross_pow_ele** = gross margin on power subscription
21. **margin_net_pow_ele** = net margin on power subscription
22. **nb_prod_act** = number of active products and services
23. **net_margin** = total net margin
24. **num_years_antig** = antiquity of the client (in number of years)
25. **origin_up** = code of the electricity campaign the customer first subscribed to
26. **pow_max** = subscribed power
27. **churn** = has the client churned over the next 3 monthsed over the next 3 months r the next 3 months

### price_data.csv 

1. **id** = client company identifier
2. **price_date** = reference date
3. **price_off_peak_var** = price of energy for the 1st period (off peak)
4. **price_peak_var** = price of energy for the 2nd period (peak)

# Visualization

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
churn = df_client[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

```python
consumption = df_client[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]
fig, axs = plt.subplots(nrows=1, figsize=(18, 5))
plot_distribution(consumption, 'cons_12m', axs)
```

```python
channel=df_client[['id', 'channel_sales', 'churn']]
```

```python
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1).fillna(0)
```

```python
channel_churn=(channel.div(channel.sum(axis=1),axis=0)*100).sort_values(by=[1],ascending=False)
```

```python
ax=channel_churn.plot(kind='bar',stacked=True,figsize=(8,6),rot=75)
annotate_stacked_bars(ax, textsize=14)
plt.title('Sales Channel')
plt.legend(['Retention','Churn'],loc='upper right')
plt.ylabel('Company base(%)')
plt.show()
```

```python
has_gas_counts = df_client['has_gas'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(4, 4))
plt.pie(has_gas_counts, labels=has_gas_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Has Gas')
plt.show()
```

```python
origin_up_counts = df_client['origin_up'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(4, 4))
plt.pie(origin_up_counts, labels=origin_up_counts.index, autopct='%1.1f%%')
plt.title('Distribution of code of the electricity campaign the customer first subscribed to')
plt.show()
```
